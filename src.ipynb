```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
import scipy.stats as s
from scipy.stats import binom
import pandas as pd
```

```python
coin_tosses = np.random.binomial(n = 5, p= 0.5,size = 10 ) # n is no. of coin , size = how many times 
print(coin_tosses)
freq = np.count_nonzero(coin_tosses == 2)
print(freq)
freq = np.count_nonzero(coin_tosses >= 2) # kam se kam 2 bar head // at least 
print(freq)
freq = np.count_nonzero(coin_tosses <= 2) # jyada se jyada // at most
print(freq)
rel_freq = freq/coin_tosses.shape[0]
print(rel_freq)
```

```python
n = 10
k = list(range (0,n+1))
p_k_outof_n =s.binom.pmf(k,n,0.5) #k=konsa outcome (0-n),n = no. of tosses,
print(p_k_outof_n)
df = pd.DataFrame({'k': k, 'PMF': p_k_outof_n})
sns.barplot(x='k', y='PMF', data=df, palette='Blues_d')
plt.title('Binomial Distribution PMF (Seaborn)')
plt.xlabel('Number of Heads (k)')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)
#sns.distplot(random.binomial(n = 10, p= 0.5,size = 10 ),hist = True,kde= False)
plt.show()
```
